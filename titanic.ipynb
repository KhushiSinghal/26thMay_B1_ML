```python
def missing_age_treatment(x):
    if pd.isnull(x['Age']):
        return df1['Age'][(df1['Sex']=='male')&(df1['Pclass']==1)].median()
    return x['Age']
```

```python
df1['Age']=df1[['Age','Pclass','Sex']].apply(missing_age_treatment,axis=1)
df1.info()
```

```python
df1.dropna(axis=0,inplace=True)
```

```python
df1.head()
```

```python
df1.info()
```

```python
df1['male'] = pd.get_dummies(df1['Sex'],drop_first=True)
```

```python
df1.info()
```

```python
df1=pd.concat([df1,pd.get_dummies(df1['Embarked'])],axis=1)
df1.info()
```

```python
df1.drop(['Sex','Embarked'],axis=1, inplace=True)
```

```python
df1.info()
```

```python
x = df1.drop('Survived',axis=1)
y = df1['Survived']
```

```python
from sklearn.model_selection import train_test_split
xtrain,xtest,ytrain,ytest = train_test_split(x,y,test_size=0.25,random_state=200)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
model = LogisticRegression().fit(xtrain,ytrain)
```

```python
from sklearn.metrics import confusion_matrix, classification_report
```

```python
ypred = model.predict(xtest)
```

```python
print(confusion_matrix(ytest,ypred))
print(classification_report(ytest,ypred))
```

```python
round(model.score(xtest,ytest)*100,2)
```

```python
pd.to_pickle(model,'TitanicSurvival')
```
