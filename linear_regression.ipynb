```python
x  =df[['Avg. Area Income', 'Avg. Area House Age', 'Avg. Area Number of Rooms', 'Area Population']]
y = df['Price']
```

```python
x.info()
```

```python
y.head()
```

<b1>1.Splitting the Data</b1>

```python
from sklearn.model_selection import train_test_split
```

```python
xtrain,xtest,ytrain,ytest=train_test_split(x,y,test_size=0.25,random_state=10)
```

```python
xtrain.info()
```

```python
xtest.info()
```

```python
ytrain.shape
```

```python
ytest.shape
```

```python
xtrain.head()
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

<b>3.Training the model with training Data</b>

```python
model.fit(xtrain,ytrain)
```

```python
model.coef_
```

```python
pd.DataFrame(model.coef_,columns=['coef,'],index=x.columns)
```

```python
print('%.1f%%'%(model.score(xtest, ytest) * 100))
```

```python
ypred = model.predict(xtest)
```

```python
ypred
```

```python
ytest
```

```python
from sklearn.metrics import mean_absolute_error
```

```python
print(mean_absolute_error(ytest,ypred))
```

```python
ypred[0]
```

```python
100*mean_absolute_error(ytest,ypred)/ypred[0]
```

<b>conclusion: let use this model, beacause the accuracy is more than 90%</b>
