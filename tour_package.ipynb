```python
df.drop('CustomerID',axis=1,inplace=True)
```

```python
df.head()
```

```python
for col in ['TypeofContact','Occupation','Gender','ProductPitched','MaritalStatus','Designation']:
    f = sns.countplot(x=col,data=df,)
    plt.title(col)
    for i in f.patches:
        x = i.get_x() + i.get_width()/2
        y = i.get_y() + i.get_height()    
        plt.annotate(i.get_height(),(x,y))
    plt.show()
```

```python
for col in ['Age','MonthlyIncome']:
    mean = df[col].mean()
    median = df[col].median()
    mode = df[col].mode().values[0]
    sns.distplot(df[col])
    plt.axvline(mean,color='g',ls='-',lw=2)
    plt.axvline(median,color='r',ls='--',lw=2)
    plt.axvline(mode,color='y',ls='-.',lw=2)
    plt.legend({'KDE':'','Mean %.1f'%mean:mean,'Median %.1f'%median:median,'Mode %.1f'%mode:mode,})
    plt.title(col)
    plt.show()
```

```python
x = df.drop('ProdTaken',axis = 1)
y=df['ProdTaken']
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import make_column_transformer
```

```python
ct = make_column_transformer(
    (OneHotEncoder(),['TypeofContact','Occupation','Gender','ProductPitched','MaritalStatus','Designation' ]), remainder = 'passthrough'
    )
```

```python
import numpy as np
```

```python
X = ct.fit_transform(x)
```

```python
X
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.2, random_state = 0)
```

```python
from sklearn.linear_model import LogisticRegression
model = LogisticRegression().fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test, y_pred)
cm
```

```python
print('%.1f%%'%(model.score(X_test, y_test) * 100))
```

```python
from sklearn.metrics import classification_report
```

```python
print(classification_report(y_test,y_pred))
```
